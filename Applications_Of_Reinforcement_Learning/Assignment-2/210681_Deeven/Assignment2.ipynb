```python
import pandas as pd
df = pd.read_csv("CarPrice_Assignment.csv")
s = len(df)
t = df.loc[1:0.8*s,:]
t1=df.loc[0.8*s:s,:]
a = df['carlength']
b = df['carwidth']
c = df['carheight']
d = df['enginesize']
e = df['stroke']
f = df['horsepower']
y = df['price']
print(t,t1)
```

```python
#cost function
def cost_fun(w1, w2, w3, w4, w5, w6, w):
    k = 0.0
    for i in range (1, int(0.8*s)):
        k += (w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w - y[i])**2
    return k/2*(int(0.8*s))    
```

```python
#gradient
def grad_fun1(w1, w2, w3, w4, w5, w6, w):
    k = 0.0
    for i in range (1, int(0.8*s)):
        k += (w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w - y[i])*a[i]
    return k/(int(0.8*s))
def grad_fun2(w1, w2, w3, w4, w5, w6, w):
    k = 0.0
    for i in range (1, int(0.8*s)):
        k += (w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w - y[i])*b[i]
    return k/(int(0.8*s))
def grad_fun3(w1, w2, w3, w4, w5, w6, w):
    k = 0.0
    for i in range (1, int(0.8*s)):
        k += (w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w - y[i])*c[i]
    return k/(int(0.8*s))
def grad_fun4(w1, w2, w3, w4, w5, w6, w):
    k = 0.0
    for i in range (1, int(0.8*s)):
        k += (w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w - y[i])*d[i]
    return k/(int(0.8*s))
def grad_fun5(w1, w2, w3, w4, w5, w6, w):
    k = 0.0
    for i in range (1, int(0.8*s)):
        k += (w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w - y[i])*e[i]
    return k/(int(0.8*s))
def grad_fun6(w1, w2, w3, w4, w5, w6, w):
    k = 0.0
    for i in range (1, int(0.8*s)):
        k += (w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w - y[i])*f[i]
    return k/(int(0.8*s))
def grad_fun7(w1, w2, w3, w4, w5, w6, w):
    k = 0.0
    for i in range (1, int(0.8*s)):
        k += (w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w - y[i])
    return k/(int(0.8*s))
```

```python
#training
x = 0.00001
w1 = 0
w2 = 0
w3 = 0
w4 = 0
w5 = 0
w6 = 0
w = 0

for i in range (1, 2000):
    w1 = w1 - x*grad_fun1(w1, w2, w3, w4, w5, w6, w)
    w2 = w2 - x*grad_fun2(w1, w2, w3, w4, w5, w6, w)
    w3 = w3 - x*grad_fun3(w1, w2, w3, w4, w5, w6, w)
    w4 = w4 - x*grad_fun4(w1, w2, w3, w4, w5, w6, w)
    w5 = w5 - x*grad_fun5(w1, w2, w3, w4, w5, w6, w)
    w6 = w6 - x*grad_fun6(w1, w2, w3, w4, w5, w6, w)
    w = w - x*grad_fun7(w1, w2, w3, w4, w5, w6, w)
```

```python
#testing
for i in range (165, 205):
    pred = w1*a[i] + w2*b[i] + w3*c[i] + w4*d[i] + w5*e[i] + w6*f[i] + w
    print("Actual price is ", y[i], "  Predicted price is ", pred)
    print("Accuracy is ", abs(pred - y[i]))
    
```
