```python
import pandas as pd
```

```python
df = pd.read_csv("/content/CarPrice_Assignment.csv")
df
```

```python
df.info()
```

```python
num_car = df.select_dtypes(include=['int64','float64'])
num_car  
```

```python
num_car= num_car.drop(['car_ID','symboling','curbweight','wheelbase','boreratio','compressionratio','peakrpm','citympg','highwaympg'],axis=1) 
num_car
```

```python
x = num_car.iloc[:,0:6].values

y = num_car.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split  
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred   
```

```python
y_test 
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_pred,y_test)
plt.plot(y_pred,y_test)
```

```python
df1 = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df1
```

```python
df1.plot(figsize=(20,9))
```

```python
df1.plot(figsize=(20,9),kind = 'bar')
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test,y_pred)
```
